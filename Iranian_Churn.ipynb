```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Customer Churn.csv")
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (6,4)
```

```python
sns.countplot(x="Churn", data=df)
plt.title("Distribution of Customer Churn")
plt.show()
```

 This chart shows how many customers churned (1) vs. stayed (0).
 If there are much more 0s than 1s, the data is imbalanced.


```python
sns.histplot(df["Age"], kde=True, bins=20, color='skyblue')
plt.title("Age Distribution")
plt.show()
```

Shows the age spread of customers.
Helps to understand which age ranges are most represented in the data.

```python
sns.boxplot(x="Churn", y="Subscription  Length", data=df)
plt.title("Subscription Length vs Churn")
plt.show()
```

Visualizes how long customers stay subscribed before leaving.
Customers with longer subscription lengths may be less likely to churn.

```python
sns.boxplot(x="Churn", y="Seconds of Use", data=df)
plt.title("Seconds of Use vs Churn")
plt.show()
```

Compares how much time (usage seconds) customers spend using the service.
Lower usage might indicate less engagement and a higher chance of churn.

```python
sns.boxplot(x="Churn", y="Frequency of SMS", data=df)
plt.title("Frequency of SMS vs Churn")
plt.show()
```

Shows how active customers are in sending SMS.
Customers sending fewer SMS messages might be more likely to churn.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Displays relationships between all numeric features.
Strong correlations (positive or negative) can indicate redundant or predictive variables.

```python
sns.boxplot(x="Churn", y="Customer Value", data=df)
plt.title("Customer Value vs Churn")
plt.show()
```

Helps understand if high-value customers tend to stay or leave.
Usually, high-value customers are more loyal.

```python
sns.scatterplot(x="Frequency of use", y="Distinct Called Numbers", hue="Churn", data=df, alpha=0.6)
plt.title("Frequency of Use vs Distinct Called Numbers")
plt.show()
```

Explores relationship between usage frequency and number of unique contacts.
Different patterns or clusters may reveal behavioral segments.

```python
sns.countplot(x="Age Group", hue="Churn", data=df)
plt.title("Age Group vs Churn")
plt.show()
```

Compares churn rates across different age groups.
Useful for identifying which age group is more likely to leave.

```python
sns.countplot(x="Complains", hue="Churn", data=df)
plt.title("Complains vs Churn")
plt.show()
```

Analyzes the effect of customer complaints on churn.
Usually, customers who complain are more likely to churn later.
